```python
import pandas as pd
import numpy as np
pd.set_option("display.max_rows",None)
pd.set_option("display.max_columns",None)
```

```python
a=pd.read_csv("CTG.csv")
```

```python
from sklearn.model_selection import train_test_split
a_test,a_train=train_test_split(a,test_size=0.2)
```

```python
# Oversampling
a1=a_train[a_train.NSP==2]
a2=a_train[a_train.NSP==3]
```

```python
a.shape
```

```python
a_train=pd.concat([a_train,a1,a1,a2,a2,a2])
```

```python
a_train.NSP.value_counts()
```

```python
a_train_x=a_train.iloc[:,:-1]
a_train_y=a_train.iloc[:,-1]
```

```python
a_test_x=a_test.iloc[:,:-1]
a_test_y=a_test.iloc[:,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(a_train_x,a_train_y)
```

```python
pred=dtree.predict(a_test_x)
pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(a_test_y,pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(a_test_y,pred)
```

```python
a_train,a_test=train_test_split(a,test_size=0.2)
```

```python
a_train_x=a_train.iloc[:,[0,1,2]]
a_train_y=a_train.iloc[:,-1]
```

```python
a_test_x=a_test.iloc[:,[0,1,2]]
a_test_y=a_test.iloc[:,-1]
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier(criterion='entropy',max_depth=7,min_samples_split=40,class_weight="balanced")
```

```python
dtree.fit(a_train_x,a_train_y)
```

```python
pred=dtree.predict(a_test_x)
pred
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(a_test_y,pred)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(a_test_y,pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
ran=RandomForestClassifier(n_estimators=50,criterion="entropy")
ran.fit(a_train_x,a_train_y)
```

```python
pred=ran.predict(a_test_x)
```
